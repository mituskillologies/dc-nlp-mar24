##### Read any wikipedia article and find top 5 words in it, which describes it.

```python
import urllib
```

```python
url = urllib.request.urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
```

```python
url
```

```python
text = url.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(text, 'html.parser')
```

```python
data = soup.get_text()
```

```python
print(data)
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(data)
```

```python
tokens
```

```python
from nltk import FreqDist
```

```python
freq =FreqDist(tokens)
```

```python
freq.plot(20)
```

##### Remove the punctuation

```python
tokens_wo_punct = []
for token in tokens:
    if token.isalpha():
        tokens_wo_punct.append(token)
```

```python
tokens_wo_punct
```

```python
freq =FreqDist(tokens_wo_punct)
freq.plot(20)
```

##### Remove the stopwords

```python
from nltk.corpus import stopwords
```

```python
swords = stopwords.words('english')
```

```python
tokens_wo_sw = []
for token in tokens_wo_punct:
    if token not in swords:
        tokens_wo_sw.append(token)
```

```python
tokens_wo_sw = [token.lower() for token in tokens_wo_punct if token.lower() not in swords]
```

```python
tokens_wo_sw
```

```python
freq =FreqDist(tokens_wo_sw)
freq.plot(20)
```

```python
freq.most_common(5)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
url = urllib.request.urlopen('https://en.wikipedia.org/wiki/Tiger')

text = url.read()
soup = BeautifulSoup(text, 'html.parser')
data = soup.get_text()
tokens = word_tokenize(data)

tokens_wo_punct = [token for token in tokens if token.isalpha()]
tokens_wo_sw = [token.lower() for token in tokens_wo_punct if token.lower() not in swords]
tokens_wo_stem = [ps.stem(token) for token in tokens_wo_sw]
freq =FreqDist(tokens_wo_stem)
freq.most_common(5)
```
